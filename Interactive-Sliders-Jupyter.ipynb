## Interactive Cosine Function

```python
from IPython.html.widgets import *
import matplotlib.pyplot as plt
import numpy as np
import math

%pylab inline
pylab.rcParams['figure.figsize'] = (15, 7)

pi = math.pi
t = np.arange(0.0,1.0,0.001)

def pltsin(f):
    plt.plot(t,np.cos(2*pi*t*f))
    plt.axis([0, 1, -1, 1])
    plt.title('cos(2*pi*f*t)')
    plt.show()
    
interact(pltsin,f=(1,20,1))
```

## Loading an image using OpenCV

```python
from IPython.html.widgets import *
import cv2
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
pylab.rcParams['figure.figsize'] = (15, 7)
path = './static/lenna.png'

img = cv2.imread(path)
rgbImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #color correction
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #greyscale 

plt.subplot(121),plt.imshow(rgbImg),plt.title('Lena in RGB')
plt.subplot(122),plt.title('Lena in Gray'),plt.imshow(img_gray,plt.get_cmap('gray'))
plt.show()
```

## Interactive Threshold

```python
def thres(Value):    
    ret,thresh1 = cv2.threshold(img_gray,Value,255,cv2.THRESH_BINARY)
    
    plt.imshow(thresh1,plt.get_cmap('gray'))
    plt.title("Interactive Threshold")
    plt.show()
    
interact(thres,Value=(1,255,2))
```

## Interactive Canny Edge Detection

```python
def canny(Value):   
    edges = cv2.Canny(img,Value,200)
    plt.imshow(edges,cmap = 'gray')
    plt.title('Canny Edge Detection')
    
interact(canny,Value=(1,255,2))
```
